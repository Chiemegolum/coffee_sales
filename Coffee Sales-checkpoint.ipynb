# Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set (color_codes= True)
```

## Importing the dataset

```python
coffee= pd.read_csv(r'Documents/DA Projects/Coffee/Coffee Sales.csv')
coffee.head()
```

### Dropping Date column

```python
coffee= coffee.drop(['date'], axis= 1)
coffee.head()
```

## Data exploration

```python
coffee.dtypes
```

### Changing the datatype of datetime to datetime 

```python
coffee['datetime']= pd.to_datetime(coffee['datetime'])
coffee.dtypes
```

### Finding duplicates

```python
coffee.shape
```

```python
coffee_duplicates= coffee[coffee.duplicated()]
print('number of duplicates:', coffee_duplicates.shape)
```

```python
coffee.count()
```

### Finding missing values

```python
coffee.isnull().sum()
```

```python
# Checking if cash paid is equivalent to the number of missing values in the card column

coffee['cash_type'].value_counts()
```

### Populating the missing values with 'null'

```python
coffee['card']= coffee['card'].fillna('null')

coffee.isnull().sum()
```

```python
coffee.head()
```

### Setting datetime as the index

```python
coffee.set_index('datetime', inplace= True)
coffee.head(5)
```

### Changing a column name

```python
coffee= coffee.rename(columns= {'money': 'price'})
coffee.tail()
```

### Checking for inconsistency

```python
coffee['coffee_name'].value_counts()
```

### Sorting by the index

```python
coffee= coffee.sort_index()
print(coffee)
```

### Creating new columns from datetime

```python
 # Extracting hour

coffee['hour']= coffee.index.hour               
coffee.head()
```

```python
 # Extracting weekdays and weekends

coffee['day_name']= coffee.index.day_name()
coffee['day_num']= coffee.index.weekday
coffee['day_type']= np.where(coffee['day_num'] < 5, 'weekday', 'weekend')
coffee
```

```python
# Extracting month-year

coffee['month_year']= coffee.index.strftime('%Y-%m')              
coffee.head()
```

### Exporting the cleaned data

```python
coffee.to_csv('Coffee Sales (cleaned).csv', index= True)
```

# Analysis

## At which hour does most sales occur during the year

```python
hourly_rush= coffee['hour'].value_counts().reset_index().sort_values(by= 'hour')

plt.figure(figsize=(10,5))
plt.plot(hourly_rush['hour'], hourly_rush['count'], linewidth= 4, color= 'violet')
plt.title('Number of Sales per Hour', fontsize= 15, fontweight= 'bold')
plt.xlabel('Hour')
plt.ylabel('Number of Sales')
```

## Which day type has the most sales

```python
rush= coffee['day_type'].value_counts().reset_index()
rush
```

## Month-to-month sales of the different coffee type

```python
monthly_sales= coffee.groupby(['month_year', 'coffee_name'])['price'].sum().reset_index()
monthly_sales
```

```python
monthly_sales= monthly_sales.sort_values('month_year').reset_index(drop= True)
```

```python
plt.figure(figsize= (12, 6))
sns.barplot(data= monthly_sales, x= 'month_year', y= 'price', 
            hue= 'coffee_name')
plt.title('Monthly Revenue of Different Coffee Types', fontsize= 15, fontweight= 'bold', pad= 15)
plt.xlabel('Month-Year', fontsize= 12)
plt.ylabel('Total Sales ($)', fontsize= 12)
plt.legend(title= 'Coffee Type', bbox_to_anchor= (1.05, 1), loc= 'upper left')
plt.style.use('grayscale')
plt.xticks(rotation= 45)
plt.show()
```

## Which coffee sales more

```python
no_of_coffee= coffee['coffee_name'].value_counts().reset_index()
no_of_coffee.sort_values(by= 'count', ascending= False)
```

```python
plt.figure(figsize= (11,6))
plt.barh(no_of_coffee['coffee_name'], no_of_coffee['count'], color= 'violet')
plt.title('Coffee types vs Number of Orders', fontweight= 'bold')
plt.xlabel('Coffee Name')
plt.ylabel('Number of Order')
```

### Which coffee generate the most total revenue 

```python
revenue= coffee[['coffee_name', 'price']]

revenue= revenue.groupby(['coffee_name'])['price'].sum().reset_index().sort_values(by= 'price', ascending= False)
```

```python
plt.figure(figsize=(12,6))

plt.bar(revenue['coffee_name'], revenue['price'], label= 'Total Revenue', color= 'violet')
# Chert formatting
plt.title('Coffee type vs Revenue', fontsize= 15, fontweight= 'bold', pad= 15)
plt.xlabel('Coffee name', fontsize= 12)
plt.ylabel('Total Sales ($)', fontsize= 12)
plt.xticks(rotation= 45)
plt.show
```

### What percentage of transactions are cash vs card

```python
payment_percentages= coffee['cash_type'].value_counts(normalize= True) * 100
payment_percentages.reset_index()
```

```python
# Preparing data
labels= ['card', 'cash']
sizes= ['97.6', '2.4']
colors= ['violet', 'red']

# Plotting
plt.pie( sizes, labels= labels, colors= colors, 
        autopct= '%1.1f%%',
        startangle= 90,
        pctdistance= 0.85)
# Creating the doghnut hole
centre_circle= plt.Circle((0, 0), 0.70, fc= 'white')
# Overlay the white circle onto the current chart
fig= plt.gcf()
fig.gca().add_artist(centre_circle)
# Styling
plt.title('Payment Method Distribution', fontweight= 'bold')
```

### Revenue per Month

```python
month_rev= coffee.groupby(['month_year'])['price'].sum().reset_index()
month_rev
```

```python
month_rev= month_rev.sort_values('month_year').reset_index(drop= True)
```

```python
plt.figure(figsize=(12,6))

plt.plot(month_rev['month_year'], month_rev['price'], linewidth= 4, color= 'violet')
# Chert formatting
plt.title('Revenue per Month', fontsize= 15, fontweight= 'bold', pad= 15)
plt.xlabel('Month-Year', fontsize= 12)
plt.ylabel('Total Sales ($)', fontsize= 12)
plt.xticks(rotation= 45)

plt.show()
```

## KPIs

```python
total_revenue= coffee['price'].sum()

total_orders= len(coffee)
```

```python
avg_order_value= total_revenue / total_orders

print(f"Total Revenue: ${total_revenue: .2f}")
print(f"Total Orders: {total_orders: }")
print(f"Avg Order Value: ${avg_order_value: .2f}")
```

```python
# Creating a figure for the KPIs
plt.figure(figsize= (12, 3))

# BOX 1: REVENUE
plt.subplot(1, 3, 1)
plt.axis('off')
plt.text(0.5, 0.7, 'Total Revenue', fontsize= 15, color= 'violet', ha= 'center', va= 'center')
plt.text(0.5, 0.35, f'${total_revenue: .2f}', fontsize= 25, fontweight= 'bold', color= 'violet', ha= 'center', va= 'center')   
# BOX 2: COFFEES SOLD
plt.subplot(1, 3, 2)
plt.axis('off')
plt.text(0.5, 0.7, 'Total Coffee Sold', fontsize= 15, color= 'violet', ha= 'center', va= 'center')
plt.text(0.5, 0.35, f'{total_orders: }', fontsize= 25, fontweight= 'bold', color= 'violet', ha= 'center', va= 'center')   
# BOX 3: AVG TRANSACTION
plt.subplot(1, 3, 3)
plt.axis('off')
plt.text(0.5, 0.7, 'Average Transaction Value', fontsize= 15, color= 'violet', ha= 'center', va= 'center')
plt.text(0.5, 0.35, f'${avg_order_value: .2f}', fontsize= 25, fontweight= 'bold', color= 'violet', ha= 'center', va= 'center')   
```
